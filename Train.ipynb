```python
import shutil

source_directory = 'kaggle/input/extended-oep'
destination_directory = 'kaggle/working/extended-oep'

# Copy the entire directory tree
shutil.copytree(source_directory, destination_directory)
```

```python
images_folders=['kaggle/input/extended-oep/extended_OEP_single/images', 'kaggle/working/extended-oep/extended_OEP_single/images']
labels_folders=['kaggle/input/extended-oep/extended_OEP_single/labels','kaggle/working/extended-oep/extended_OEP_single/labels']
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define source and destination folders
source_images_folders = ['kaggle/input/extended-oep/extended_OEP_single/images', 'kaggle/working/extended-oep/extended_OEP_single/images']

source_labels_folders = ['kaggle/input/extended-oep/extended_OEP_single/labels','kaggle/working/extended-oep/extended_OEP_single/labels']

destination_folder = 'kaggle/working'

# Create destination folders for train and validation sets
os.makedirs(os.path.join(destination_folder, 'images', 'train'), exist_ok=True)
os.makedirs(os.path.join(destination_folder, 'images', 'val'), exist_ok=True)
os.makedirs(os.path.join(destination_folder, 'labels', 'train'), exist_ok=True)
os.makedirs(os.path.join(destination_folder, 'labels', 'val'), exist_ok=True)

# Combine images into one folder
all_images = []
all_labels = []

for src_images, src_labels in zip(source_images_folders, source_labels_folders):
    images_list = os.listdir(src_images)
    labels_list = os.listdir(src_labels)
    all_images.extend([(os.path.join(src_images, img), img) for img in images_list])
    all_labels.extend([(os.path.join(src_labels, label), label) for label in labels_list])

# Split data into train and validation sets
images_train, images_val, labels_train, labels_val = train_test_split(
    all_images, all_labels, test_size=0.15, random_state=42
)

# Move images to destination folders
for src, dest in [(images_train, 'train'), (images_val, 'val')]:
    for src_path, src_name in src:
        dest_path = os.path.join(destination_folder, 'images', dest, src_name)
        shutil.copy(src_path, dest_path)

# Move labels to destination folders
for src, dest in [(labels_train, 'train'), (labels_val, 'val')]:
    for src_path, src_name in src:
        dest_path = os.path.join(destination_folder, 'labels', dest, src_name)
        shutil.copy(src_path, dest_path)

print("Data has been successfully split and organized.")
```
